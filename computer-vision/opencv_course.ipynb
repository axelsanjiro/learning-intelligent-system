```python
import cv2
import os
```

```python
# An image shape is given as (height, width, channels)
# in "most cases" pixel value range from 0 to 255.
# in binary images, pixel values are either [0, 1] (or[0, 255].
# in 16 bits images, pixel values range from 0 to 65535.
```

```python

# read image
image_path = os.path.join('..', 'datasets', 'comvis', 'art.jpg')

img = cv2.imread(image_path)
img = cv2.resize(img, (600, 600))  # resize image to width=600, height=400
```

```python
# write image
output_path = os.path.join('..', 'datasets', 'comvis', 'art_out.jpg')
cv2.imwrite(output_path, img)
```

```python
# Visualize image
cv2.imshow('image', img)
cv2.waitKey(0)
```

```python
# Read video
video_path = os.path.join('..', 'datasets', 'comvis', 'squirel.mp4')

video = cv2.VideoCapture(video_path)

```

```python
# Visualize Video
ret = True
while ret:
    ret, frame = video.read()

    if ret:
        cv2.imshow('video', frame)
        if cv2.waitKey(40) & 0xFF == ord('q'):
            break

video.release()
cv2.destroyAllWindows()
```

```python
# Read webcam
webcam = cv2.VideoCapture(0)
```

```python
# Visualize webcam

while True:
    ret, frame = webcam.read()

    cv2.imshow('webcam', frame)
    if cv2.waitKey(40) & 0xFF == ord('q'): #  Press 'q' to  exit
        break


webcam.release()
cv2.destroyAllWindows()
```

```python
# resizing
import os 
import cv2 
```

```python
img = cv2.imread(os.path.join('..', 'datasets', 'comvis', 'art.jpg'))

resize_img = cv2.resize(img, (600,600))

print(img.shape)  # original shape
print(resize_img.shape)  # resized shape

cv2.imshow('original', img)
cv2.imshow('resized', resize_img)
cv2.waitKey(0)
```

```python
# cropping
img = cv2.imread(os.path.join('..', 'datasets', 'comvis', 'art.jpg'))
img = cv2.resize(img, (600,600))
print(img.shape)
cropped_img = img[100:600, 100:600]

cv2.imshow('original', img)
cv2.imshow('cropped', cropped_img)
cv2.waitKey(0)
```
